```python
def MakeContColPlots(df,contCols):
    title_specs = {'fontsize': 9, 'fontweight': 'bold', 'color': '#992600'}

    grid_specs = {'visible': True, 'which': 'both', 'linestyle': '--', 
                           'color': 'lightgrey', 'linewidth': 0.75}
    
    fig, axes = plt.subplots(len(contCols), 2 ,figsize = (16, len(contCols) * 4.2), 
                                    gridspec_kw = {'hspace': 0.35, 
                                                'wspace': 0.3, 
                                                'width_ratios': [0.80, 0.20]
                                                }
                                )
    
    for i,col in enumerate(contCols):
            ax = axes[i,0]
            sns.kdeplot(data = df, x = col, 
                        palette = ['#0039e6', '#ff5500', '#00b300'],hue=target,
                        ax = ax, linewidth = 2.1
                        )
            ax.set_title(f"\n{col}", **title_specs)
            ax.grid(**grid_specs)
            ax.set(xlabel = '', ylabel = '')

            ax = axes[i,1]
            sns.boxplot(data = df, y = col, width = 0.25,
                        color = '#33ccff', saturation = 0.90, linewidth = 0.90, 
                        fliersize= 2.25,
                        ax = ax)
            ax.set(xlabel = '', ylabel = '')
            ax.set_title(col, **title_specs)
            
    plt.suptitle(f"\nDistribution analysis- continuous columns\n", **title_specs, 
                        y = 0.92, x = 0.50
                    )
    plt.tight_layout()
    plt.show()
```
